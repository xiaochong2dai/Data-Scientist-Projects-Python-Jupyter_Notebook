```python
AGE=np.array(df['AGE'].values)
CRIM=np.array(df['CRIM'].values)
MEDV=np.array(df['MEDV'].values)
```

```python
# lazy way
plt.figure(figsize=(20,5))

plt.subplot(131)
plt.hist(df['AGE'],25,alpha=.5)
plt.xlabel('AGE')
plt.ylabel('Count')
plt.title('Age Distribution')

plt.subplot(132)
plt.hist(df['CRIM'],25,color='r',alpha=.5)
plt.xlabel('CRIM')
plt.ylabel('Count')
plt.title('CRIME Distribution')

plt.subplot(133)
plt.hist(df['MEDV'],25,color='g',alpha=.5)
plt.xlabel('MEDV')
plt.ylabel('Count')
plt.title('MEDV Distribution')

```

```python
# better way

fig = plt.figure(figsize=(20,5))

ax1 = fig.add_subplot(1,3,1)
ax1.hist(AGE, bins=25, alpha=.3)
ax1.set_xlabel('AGE')
ax1.set_title('Age Distribution')

ax2 = fig.add_subplot(1,3,2)
ax2.hist(CRIM, bins=25, color='r', alpha=.5)
ax2.set_xlabel('CRIM')
ax2.set_title('CRIM Distribution')

ax3 = fig.add_subplot(1,3,3)
ax3.hist(MEDV, bins=25, color='g',alpha=.5)
ax3.set_xlabel('MEDV')
ax3.set_title('MEDV Distribution')

fig.show()
```

#### Q4. show the relationship bewteen MEDV and AGE and CHAS?

```python
redDots = df[df['CHAS']==0][['MEDV','AGE']]
blueDots = df[df['CHAS']==1][['MEDV','AGE']]

image = plt.figure(figsize=(10,5))
plt.subplot(121)
plt.scatter(redDots['AGE'],redDots['MEDV'],c='r',alpha=.5)
plt.xlabel('AGE')
plt.ylabel('MEDV')
plt.title('CHAS = 0')

plt.subplot(122)
plt.scatter(blueDots['AGE'],blueDots['MEDV'],c='b',alpha=.5)
plt.xlabel('AGE')
plt.ylabel('MEDV')
plt.title('CHAS = 1')
```
